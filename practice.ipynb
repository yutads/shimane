```python
from sklearn.linear_model import LinearRegression
x = np.linspace(0, 10, 1000)
y = (0.8)*x + 2.5 * np.random.randn(1000) + 25
#y = (0.8)*x + np.random.randn(1000) + 25

lr = LinearRegression()
lr.fit(np.array([x]).T,y)

plt.figure(figsize=(6, 6), dpi=80)
plt.scatter(x,y)
plt.grid()                     # グリッド線を表示
plt.show()                     # 図の表示

print('切片: ', lr.intercept_)
print('傾き: ', lr.coef_[0])
```
